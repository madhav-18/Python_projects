```python
#Number Guessing Game Objectives:
import random

logo = ''' _______    __  __   ______   ______   ______       _________  ___   ___   ______       _______    ________   ___ __ __   ______      
/______/\  /_/\/_/\ /_____/\ /_____/\ /_____/\     /________/\/__/\ /__/\ /_____/\     /______/\  /_______/\ /__//_//_/\ /_____/\     
\::::__\/__\:\ \:\ \\::::_\/_\::::_\/_\::::_\/_    \__.::.__\/\::\ \\  \ \\::::_\/_    \::::__\/__\::: _  \ \\::\| \| \ \\::::_\/_    
 \:\ /____/\\:\ \:\ \\:\/___/\\:\/___/\\:\/___/\      \::\ \   \::\/_\ .\ \\:\/___/\    \:\ /____/\\::(_)  \ \\:.      \ \\:\/___/\   
  \:\\_  _\/ \:\ \:\ \\::___\/_\_::._\:\\_::._\:\      \::\ \   \:: ___::\ \\::___\/_    \:\\_  _\/ \:: __  \ \\:.\-/\  \ \\::___\/_  
   \:\_\ \ \  \:\_\:\ \\:\____/\ /____\:\ /____\:\      \::\ \   \: \ \\::\ \\:\____/\    \:\_\ \ \  \:.\ \  \ \\. \  \  \ \\:\____/\ 
    \_____\/   \_____\/ \_____\/ \_____\/ \_____\/       \__\/    \__\/ \::\/ \_____\/     \_____\/   \__\/\__\/ \__\/ \__\/ \_____\/ 
                                                                                                                                      '''

EASY_LEVEL_ATTEMPT = 10
HARD_LEVEL_ATTEMPT = 5
EXTRA_HARD_LEVEL_ATTEMPT = 3

def set_difficulty():
    """returns no. of attemps in selected difficulty"""
    level = input("choose a difficulty. Type 'easy' or 'hard' or 'extra hard': ").lower()
    if level == "easy":
        return EASY_LEVEL_ATTEMPT
    elif level == "hard":
        return HARD_LEVEL_ATTEMPT
    elif level == "extra hard":
        return EXTRA_HARD_LEVEL_ATTEMPT

def hint(guess, answer ,turns):
    """Checks the guess with answer,if wrong returns no. of turns left."""
    if guess > answer :
        print("Too high.")
        return turns - 1
    if guess < answer :
        print("Too low.")
        return turns - 1
    else:
        print(f"You got it. The answer was {answer}")

        
def game():  
    print(logo)
    print("Welcome to the Number Guessing Game!")
    print("I am thinkong of a number between 1 to 100")
    answer = random.randint(1,100) 
    # print(f"Psst.. The chosen number is {answer}")
    
    turns = set_difficulty()
    #Repeat the guessing functionality if they get it wrong.
    guess = 0
    
    while guess != answer :
        print(f"You have {turns} attempts remaining to guess the number.")
        """takes guess from user."""
        guess = int(input("Make a guess: "))
        
        turns = hint(guess, answer, turns)
        
        """prints how many turns does user has left"""
        if turns == 0:
            print("You ran out of attempts. You lose!")
            print(f"The answer was {answer}")
            return 
        elif guess != answer:
            print("Guess again")

game()



```
